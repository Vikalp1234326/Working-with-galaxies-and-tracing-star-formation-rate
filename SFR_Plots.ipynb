### All the points for all the tracers in one plot

```python
import matplotlib.pyplot as plt

# Create a 1x2 grid
fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# ---------------------------------------------------
# Top Panel: ΣSFR comparison (Paα, 24μm+Hα, Leroy+13, VLA, VENGA Hα & Hβ)
# ---------------------------------------------------
ax = axes[0]

ax.plot(rgal_kpc, sig_SFR_kpc2, "o", ms=1.5, c="red", label=r"Pa$\alpha$ (JWST)")
ax.plot(rgal_kpc, uc_sfr, "s", ms=1.5, c="blue", label=r"24μm + Hα")
ax.plot(rgal_kpc, sigma_SFR, "h", ms=1.5, c="green", label=r"Leroy et al. (2013)")
ax.plot(rgal_kpc, sigma_sfr_vla, "*", ms=1.9, c="orange", label="33 GHz VLA")
#ax.plot(rgal_kpc, sigma_sfr_venga_ha, "D", ms=1.7, c="orange", label="VENGA Hα")
ax.plot(rgal_kpc, sigma_sfr_venga_extinction_corr, "v", ms=1.7, c="brown", label="VENGA Hα Extinction Corrected")

ax.set_yscale("log")
ax.set_xlabel("Galactocentric Radius [kpc]", fontweight='bold', fontsize=14)
ax.set_ylabel(r"$\Sigma_{\mathrm{SFR}}$ [M$_{\odot}$ yr$^{-1}$ kpc$^{-2}$]", fontweight='bold', fontsize=14)
ax.set_title(r"Comparison of $\Sigma_{\mathrm{SFR}}$", fontweight='bold', fontsize=16)
ax.legend(fontsize=10)
ax.grid(alpha=0.3, linestyle=':')

# ---------------------------------------------------
# Bottom Panel: SFR comparison (Paα, 24μm+Hα, Leroy+13, VLA, VENGA Hα & Hβ)
# ---------------------------------------------------
ax = axes[1]

ax.plot(rgal_kpc, SFR_array, "o", ms=1.5, c="red", label=r"Pa$\alpha$ (JWST)")
ax.plot(rgal_kpc, sfr_value, "s", ms=1.5, c="blue", label=r"24μm + Hα")
ax.plot(rgal_kpc, SFR_from_sigma, "h", ms=2.5, c="green", label=r"Leroy et al. (2013)")
ax.plot(rgal_kpc, sfr_vla, "*", ms=1.5, c="orange", label="33 GHz VLA")
#ax.plot(rgal_kpc, sfr_venga_ha, "D", ms=2.7, c="orange", label="VENGA Hα", alpha = 0.3)
ax.plot(rgal_kpc, sfr_venga_extinction_corr, "v", ms=1.7, c="brown", label="VENGA Hα Extinction Corrected")

ax.set_yscale("log")
ax.set_xlabel("Galactocentric Radius [kpc]", fontweight='bold', fontsize=14)
ax.set_ylabel(r"SFR [M$_{\odot}$ yr$^{-1}$]", fontweight='bold', fontsize=14)
ax.set_title(r"Comparison of SFR", fontweight='bold', fontsize=16)
ax.legend(fontsize=10, borderaxespad=0.5)
ax.grid(alpha=0.3, linestyle=':')

# ---------------------------------------------------
# Adjust layout and save
# ---------------------------------------------------
plt.tight_layout()
plt.show()


```
